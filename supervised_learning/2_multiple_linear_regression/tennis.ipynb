# Tennis Ace

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
# Load data into dataframe
tennis_df = pd.read_csv('tennis_stats.csv')
```

```python
# Explore different features relationships
feat_1 = tennis_df[['Aces']]
feat_2 = tennis_df[['ServiceGamesPlayed']]
plt.scatter(feat_1,feat_2)
plt.show()
```

```python
# Explore different features relationships
feat_3 = tennis_df[['FirstServePointsWon']]
feat_4 = tennis_df[['TotalServicePointsWon']]
plt.scatter(feat_3,feat_4)
plt.show()
```

```python
# Explore different features relationships
feat_5 = tennis_df[['BreakPointsOpportunities']]
feat_6 = tennis_df[['Winnings']]
plt.scatter(feat_5,feat_6)
plt.show()
```

Let's fit a **single linear regression model** to 'BreakPointsOpportunities' and 'Winnings' features 

```python
# Split test/train sets
feat_train, feat_test, outcome_train, outcome_test = train_test_split(feat_5,feat_6,train_size=0.8)
```

```python
# Create simple linear model
sl = LinearRegression()
sl.fit(feat_train,outcome_train)
print(sl.score(feat_test,outcome_test))

outcome_predict = sl.predict(feat_test)
plt.scatter(feat_5,feat_6)
plt.scatter(feat_test,outcome_predict)
plt.show()
```

Let's fit a **multiple linear regression model** to 'BreakPointsOpportunities', 'FirstServeReturnPointsWon' and 'Winnings' features 

```python
features = tennis_df[['BreakPointsOpportunities',
'FirstServeReturnPointsWon']]
outcome = tennis_df[['Winnings']]

# Split test/train sets
feat_train, feat_test, outcome_train, outcome_test = train_test_split(features, outcome,train_size=0.8)
```

```python
# Create multiple linear model
ml = LinearRegression()
ml.fit(feat_train,outcome_train)
print(ml.score(feat_test,outcome_test))
print(feat_test)
outcome_predict = ml.predict(feat_test)
```

```python
# Plot BreakPointsOpportunities vs outcome prediction
plt.scatter(features['BreakPointsOpportunities'],outcome)
plt.scatter(feat_test['BreakPointsOpportunities'],outcome_predict)
plt.show()
```

```python
# Plot FirstServeReturnPointsWon vs outcome prediction
plt.scatter(features['FirstServeReturnPointsWon'],outcome)
plt.scatter(feat_test['FirstServeReturnPointsWon'],outcome_predict)
plt.show()
```
